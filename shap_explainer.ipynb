```python
import pandas as pd
import shap
import joblib

# Load the trained model
model = joblib.load("models/xgb_selected_features_model.pkl")

# Load the engineered dataset
df = pd.read_csv("data/interest_rate_df_engineered.csv")

# Drop the target column
X = df.drop(columns=["int_rate"])

# Create SHAP explainer
explainer = shap.Explainer(model)

# Save explainer to the app folder
joblib.dump(explainer, "app/shap_explainer.pkl")
```
